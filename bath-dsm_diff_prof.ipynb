plot channel profiles to show slope from 2016 DSM, along with channel profiles from bathymetric slope from 2000 and meander belt axis profiles from bathymetry

```python
# read csv files

import numpy as np
import pandas as pd

# dataframe for bathymetric profiles
df1 = pd.read_csv('bath_profs_shore.csv', header=0)

# dataframe for profiles of difference between DSM16 and bathymetry
df2 = pd.read_csv('dsm16-bath_diff_profs.csv', header=0)
```

```python
# bathymetric profile data

A = df1['prof_id']== 1
p1_bath = df1[A]

B = df1['prof_id']== 2
p2_bath = df1[B]

C = df1['prof_id']== 3
p3_bath = df1[C]

# dsm-bath difference profiles

D = df2['prof_id']== 1
p1_diff = df2[D]

E = df2['prof_id']== 2
p2_diff = df2[E]

F = df2['prof_id']== 3
p3_diff = df2[F]
```

```python
# plot data

%matplotlib inline 
from matplotlib import pyplot as plt

fig = plt.figure(figsize = (12,8))

ax1 = plt.subplot(2,1,1)
ax1.plot(p1_bath.cds2d, p1_bath.bathymetry, '--', label='profile 1')
ax1.plot(p2_bath.cds2d, p2_bath.bathymetry, '--', label='profile 2')
ax1.plot(p3_bath.cds2d, p3_bath.bathymetry, '--', label='profile 3')

plt.legend(loc='upper left',fontsize=12)
plt.xlim(1600,0)
#plt.xlabel('distance [m]',fontsize=12)
plt.ylabel('elevation [m]',fontsize=14)
plt.tick_params(axis='both', which='major', labelsize=12)
#plt.title('Evolution of longitudinal topography, CM1', fontsize=14)

ax2 = plt.subplot(2,1,2)
ax2.plot(p1_diff.cds2d, p1_diff.DSM16_bath, '--', label='profile A')
ax2.plot(p2_diff.cds2d, p2_diff.DSM16_bath, '--', label='profile B')
ax2.plot(p3_diff.cds2d, p3_diff.DSM16_bath, '--', label='profile C')
plt.text(50,-8,'E',fontsize=16)
plt.text(1570,-8,'W',fontsize=16)

plt.legend(loc='upper left',fontsize=12)
plt.xlim(1600,0)
plt.xlabel('distance [m]',fontsize=14)
plt.ylabel('difference [m]',fontsize=14)
plt.tick_params(axis='both', which='major', labelsize=12)

# save figure
plt.savefig('bath-prof_dsm-bath_diff_prof.png', bbox_inches='tight')
#plt.savefig('CM1_channel-slope-evo.svg')
```
